```python
# Setting environment variables in the notebook
lr = "3e-5"
c = "0.6"
s = "100"

# Display the variables
print(lr)

# Setting up the MODEL_DIR variable
MODEL_DIR = f"Model_Hackathon/JointBERT-CRF_PhoBERTencoder/{lr}/{c}/{s}"
print(MODEL_DIR)

# Running the python script using the converted bash command
!python3 main.py --token_level word-level \
                  --model_type phobert \
                  --model_dir $MODEL_DIR \
                  --data_dir Hackathon \
                  --seed $s \
                  --do_train \
                  --do_eval \
                  --save_steps 56 \
                  --logging_steps 56 \
                  --num_train_epochs 50 \
                  --tuning_metric mean_intent_slot \
                  --use_crf \
                  --gpu_id 0 \
                  --embedding_type soft \
                  --intent_loss_coef $c \
                  --learning_rate $lr
```

```python
# First, run the previous bash script to train JointBERT-CRF_PhoBERTencoder
# !bash run_jointBERT-CRF_PhoBERTencoder.sh

# Now set up environment variables for the new training run
lr = "4e-5"
c = "0.5"
s = "100"

# Display the variables
print(lr)

# Set the MODEL_DIR path for JointIDSF
MODEL_DIR = f"Model_Hackathon/JointIDSF_PhoBERTencoder/{lr}/{c}/{s}"
print(MODEL_DIR)

# Run the second python script using the pretrained model path
!python3 main.py --token_level word-level \
                  --model_type phobert \
                  --model_dir $MODEL_DIR \
                  --data_dir Hackathon \
                  --seed $s \
                  --do_train \
                  --do_eval \
                  --save_steps 56 \
                  --logging_steps 56 \
                  --num_train_epochs 50 \
                  --tuning_metric mean_intent_slot \
                  --use_intent_context_attention \
                  --attention_embedding_size 200 \
                  --use_crf \
                  --gpu_id 0 \
                  --embedding_type soft \
                  --intent_loss_coef $c \
                  --pretrained \
                  --pretrained_path Model_Hackathon/JointBERT-CRF_PhoBERTencoder/3e-5/0.6/100 \
                  --learning_rate $lr
```
