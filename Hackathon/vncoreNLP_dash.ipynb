```python
with open(syllable_train_seq_out, 'r', encoding= 'utf-8') as file:
  labels = file.readlines()
  file.close()
```

```python
def mergeDash(line, new_line, label, id):
  words = line.split()
  new_words = new_line.split()
  res_words = []
  slots = label.split()
  new_slots = []

  ptr = 0
  errflag = False

  for word in new_words:
    if ptr >= len(slots):
      break
    if '_' in word:
      sub_word = word.split('_')
      composes = slots[ptr : ptr + len(sub_word)]
      composes = list(set(composes))
      if len(composes) != 1:
        types = [x.split('-')[-1] for x in composes]
        if len(list(set(types))) != 1:
          print(id, word, slots[ptr : ptr + len(sub_word)])
          errflag = True
          res_words.extend(sub_word)
          new_slots.extend(slots[ptr : ptr + len(sub_word)])
        else:
          res_words.append(word)
          new_slots.append(slots[ptr])
      else:
        res_words.append(word)
        new_slots.append(slots[ptr])
      ptr += len(sub_word)
    else:
      new_slots.append(slots[ptr])
      res_words.append(word)
      ptr += 1

  return new_slots, errflag, ' '.join(res_words)

id = 0
mergeDash(lines[id], new_lines[id], labels[id], id)
```

```python
len(lines), len(new_lines), len(labels)
```

```python
new_labels = []
res_lines = []
errid = []
import random as rd
arr = list(range(len(lines)))
# rd.shuffle(arr)
for id in arr:
  new_slots, errflag, res_line = mergeDash(lines[id], new_lines[id], labels[id], id)
  if errflag:
    errid.append(id)
  new_labels.append(' '.join(new_slots))
  res_lines.append(res_line)
```
